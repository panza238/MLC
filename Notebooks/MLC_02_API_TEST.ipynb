## Armado del Dataset
10 categorías --> 1000 registros por categoría

```python
my_cat_req = requests.get("https://api.mercadolibre.com/sites/MLA/search?category=MLA1648")
#https://api.mercadolibre.com/sites/MLA/search?category=MLA5726
```

```python
my_cat_req.json()['results']
```

```python
categories = {
    "computacion" : "MLA1648",
    "electrodomesticos" : "MLA5726",
    "deportes" : "MLA1276",
    "belleza" : "MLA1246",
    "juegos" : "MLA1132",
    "celulares" : "MLA1051",
    "electronica" : "MLA1000",
    "ropa" : "MLA1430",
    "hogar" : "MLA1574",
    "consolas" : "MLA1144"}
```

```python
categories.values()
```

```python
# TODO: Here
req_url = "https://api.mercadolibre.com/sites/MLA/search?category="

first_pass = True  # Flag para inicializar el df.

for cat_value in categories.values():
    i = 0
    
    while i <= 1000:
        request_string = req_url + cat_value + "&offset={}".format(i)
        request = requests.get(request_string)
        results = request.json()["results"]
        print(cat_value, i, request_string)
        
        if first_pass:  # Inicializo el df 
            results_df = pd.json_normalize(results)
            first_pass = False  # toggle flag
            print("FIRST PASS") 
        
        else:  # append to existing df
            df_to_append = pd.json_normalize(results)
            results_df = results_df.append(df_to_append, ignore_index=True)
        
        i += 50
    
```
